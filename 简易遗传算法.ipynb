```python
import numpy as np

m=4                 #种群规模
p_c=1.0             #交叉概率
p_m=0.05            #变异概率
n=100                 #总次数
group=[0]*m
group_new=[0]*m
code1=[]
length=5
for i in range(m):
    group[i]=np.random.randint(0,32)
print(group)

def binary(a,length):
    b=[]
    while 1:
        s=a//2
        y=a%2
        b=b+[y]
        if s==0:
            break
        a=s
    while len(b)<length:
        b.append('0')
    b.reverse()
    b = [ str(i) for i in b ]
    b=''.join(b)
    return b
    
    
def translate(group,length):
    code=[]
    for i in range(m):
        code.append(binary(group[i],length))
    return code

def decode(code):
    group=[]
    for i in range(m):
        group.append(int(code[i],2))
    return group

    

def fit_x(a):
    fit=a**2
    return fit

def probability(group):
    P=[]
    total_fit=0
    for i in range(m):
        total_fit+=fit_x(group[i])
    for i in range(m):
        P.append(fit_x(group[i])/total_fit)
    return P

def pro_plus(P):
    P_p=[]
    Pplus=0
    for i in range(m):
        Pplus+=P[i]
        P_p.append(Pplus)
    return P_p

def select(group):        
    P=probability(group)
    Pp=pro_plus(P)
    ran=[0]*m
    group_new=[0]*m
    for i in range(m):
        ran[i]=np.random.rand()
        if ran[i]<Pp[0]:
            group_new[i]=group[0]
        elif Pp[0]<ran[i]<Pp[1]:
            group_new[i]=group[1]
        elif Pp[1]<ran[i]<Pp[2]:
            group_new[i]=group[2]
        else:
            group_new[i]=group[3]
    return group_new

def change(code,p_c):
    a=0
    b=0
    c=0
    a=np.random.randint(1,4)
    sp=np.random.randint(1,3)
    b=a
    c=a
    ran=np.random.rand()
    for i in range(m):
        code[i]=str(code[i])
    while(a==b):
        b=np.random.randint(1,4)
    while(a==c) or (b==c):   
        c=np.random.randint(1,4)  
    code1=[]
    if ran<p_c:
        temp1=[]
        temp2=[]
        temp1.extend(code[0][0:sp])
        temp1.extend(code[a][sp:length])
        temp2.extend(code[a][0:sp])
        temp2.extend(code[0][sp:length])
        code1.append(temp1)
        code1.append(temp2)
        temp1=[]
        temp2=[]
        temp1.extend(code[b][0:sp])
        temp1.extend(code[c][sp:length])
        temp2.extend(code[c][0:sp])
        temp2.extend(code[b][sp:length])
        code1.append(temp1)
        code1.append(temp2)
    return code1

def mutation(code,p_m):
    for i in range(m):
        random=np.random.rand()
        if random<p_m:
            s=np.random.randint(0,length)
            if code[i][s]=='0':
                code[i][s]='1'
            else:
                code[i][s]='0'
    for i in range(m):
            code[i]=''.join(code[i])
    return code
```

```python
def inherit(group,n,p_c,p_m):
    code1=[0]*m
    maxfit=0
    for i in range(n):
        group_new=select(group)
        code=translate(group_new,length)
        print(code)
        code1=change(code,p_c)
        code1=mutation(code1,p_m)
        group=decode(code1)
        for i in range(m):
            if fit_x(group[i])>maxfit:
                maxfit=fit_x(group[i])
        if maxfit==961:
            break
    return maxfit
print(inherit(group,n,p_c,p_m))
        

        
        
        
        
```
