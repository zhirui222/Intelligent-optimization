```python
import numpy as np

m=50                #种群规模
p_c=0.75             #交叉概率
p_m=0.05            #变异概率
n=200                 #总次数
group=[0]*m
group_new=[0]*m
code1=[]
length=22
mh=int(m/2)
mutation_num=1
sample=['']*m
group=np.random.randint(0,2,(50,22))                        #生成一个0-1随机种群
for i in range(m):
    for j in range(length): 
        sample[i]+=str(group[i][j])

def binary(a,length):                                        #二进制
    b=[]
    while 1:
        s=a//2
        y=int(a%2)
        b=b+[y]
        if s==0:
            break
        a=s
    while len(b)<length:
        b.append('0')
    b.reverse()
    b = [ str(i) for i in b ]
    b=''.join(b)
    return b
    
    
def translate(group,length):                                #翻译
    code=[]
    value=0
    for i in range(m):
        value=(group[i]+1)*(2**22-1)/3
        code.append(binary(value,length))
    return code

def decode(code):                                            #解码
    group=[]
    value=0
    for i in range(m):
        value=(int(code[i],2))/(2**22-1)*3-1
        group.append(value)                           
    return group
cc=decode(sample)
print(cc)




    

def fit_x(a):                                               #适应度函数
    fit=a*np.sin(10*np.pi*a)+2.0
    return fit

def probability(group):                                   #可能性
    P=[]
    total_fit=0
    for i in range(m):
        total_fit+=fit_x(group[i])
    for i in range(m):
        P.append(fit_x(group[i])/total_fit)
    return P

def pro_plus(P):                             #可能性叠加
    P_p=[]
    Pplus=0
    for i in range(m):
        Pplus+=P[i]
        P_p.append(Pplus)
    return P_p

def select(group):                            #从种群里选
    P=probability(group)
    Pp=pro_plus(P)
    ran=[0]*m
    group_new=[0]*m
    for i in range(m):
        ran[i]=np.random.rand()
        for j in range(m):
            if ran[i]<Pp[j]:
                group_new[i]=group[j]
                break
    return group_new

def change(code,p_c):                            #配对交叉函数                                               
    sp=np.random.randint(0,21)                   #顺序生成
    ran=np.random.rand()
    code1=[]
    for j in range(m):
        code[j]=str(code[j])
    for i in range(mh):
        if ran<p_c:
            temp1=[]
            temp2=[]
            temp1.extend(code[2*i][0:sp])
            temp1.extend(code[2*i+1][sp:length])
            temp2.extend(code[2*i+1][0:sp])
            temp2.extend(code[2*i][sp:length])
            code1.append(temp1)
            code1.append(temp2)
        else:
            temp1=[]
            temp2=[]
            temp1.extend(code[2*i][:])
            temp2.extend(code[2*i+1][:])
            code1.append(temp1)
            code1.append(temp2)       
    return code1

def mutation(code,p_m):
    for i in range(m):
        random=np.random.rand()
        if random<p_m:
            s=np.random.randint(0,length)
            if code[i][s]=='0':
                code[i][s]='1'
            else:
                code[i][s]='0'
    for i in range(m):
            code[i]=''.join(code[i])
    return code
```

```python
def inherit(group,n,p_c,p_m):
    code1=[0]*m
    maxfit=0
    maxnum=0
    for i in range(n):
        group_new=select(group)
        code=translate(group_new,length)
        code1=change(code,p_c)
        code1=mutation(code1,p_m)
        group=decode(code1)
        for i in range(m):
            if fit_x(group[i])>maxfit:
                maxfit=fit_x(group[i])
                maxnum=group[i]
    return maxfit,maxnum
print(inherit(cc,n,p_c,p_m))
        
```
