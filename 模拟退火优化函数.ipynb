```python
import numpy as np
import matplotlib.pyplot as plt

def arrayturn(a):
    b=[0,0]
    for i in range(2):
        b[i]=a[i]
    return b


L=100
a=0.99
k=0.99
s=0.01
T=100
YZ=1E-8
rmax=5
rmin=-5
x_prebest=[]
x_best=[]
P=0
valuelist=[]
for i in range(2):
    x_prebest.append(np.random.rand()*(rmax-rmin)+rmin)
    x_best.append(np.random.rand()*(rmax-rmin)+rmin)
x_old=[]
x_old=arrayturn(x_prebest)
x_new=arrayturn(x_best)
print(s*(2*(np.random.rand()-0.5)*(rmax-rmin)+rmin))
def function(x):
    value=5*np.cos(x[0]*x[1])+x[0]*x[1]+x[1]*x[1]*x[1]
    return value
    
```

```python
while (T>0.01) and (abs(function(x_new)-function(x_prebest))>YZ):
    T=a*T
    for i in range(L):
        for i in range(2):
            x_new[i]=x_old[i]+s*(2*(np.random.rand()-0.5)*(rmax-rmin))                      #产生一个新解
            if (x_new[i]>rmax):                                                             #判断边界
                x_new[i]=rmax
            elif(x_new[i]<rmin):                                                                                  
                x_new[i]=rmin
        if function(x_best)>function(x_new):
            valuelist.append(function(x_prebest))
            x_prebest=arrayturn(x_best)
            x_best=arrayturn(x_new)
            P+=1
        if (function(x_old)>function(x_new)):
            x_old=arrayturn(x_new)
        else:
            if np.exp(-(function(x_new)-function(x_old))/T)>np.random.rand():
                x_old=arrayturn(x_new)                
print(x_old)
print(function(x_old))
```

```python
fig=plt.figure(num=1,figsize=(10,10))
x=np.arange(1,P+1)
plt.plot(x,valuelist)
plt.show
```
